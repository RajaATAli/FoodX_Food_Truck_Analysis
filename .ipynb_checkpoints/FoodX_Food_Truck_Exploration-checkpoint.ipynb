```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns


# Load the CSV data
food_trucks = pd.read_csv('food_trucks.csv')
```

```python
# Display the first few rows of the DataFrame
print(food_trucks.head())

# Display basic statistics for the numeric columns (e.g., Rating)
print(food_trucks.describe())
```

## Data Cleaning

```python
# Function to calculate the difference in hours between two times
def time_difference(start_time, end_time):
    if start_time is None or end_time is None:
        return 0
    try:
        time_format = "%I:%M %p"
        start_dt = datetime.strptime(start_time, time_format)
        end_dt = datetime.strptime(end_time, time_format)
        diff = end_dt - start_dt
        hours = diff.seconds / 3600
        return hours
    except:
        return 0

# Splitting the days and times for each food truck
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Initialize new columns for each day's opening and closing times
for day in days:
    food_trucks[f'{day}_Open'] = None
    food_trucks[f'{day}_Close'] = None

# Update the new columns based on the Opening Hours information
for index, row in food_trucks.iterrows():
    if pd.notna(row['Opening Hours']):
        for day in days:
            if day in row['Opening Hours']:
                day_timing = row['Opening Hours'].split(f'{day}: ')[1].split('\n')[0]
                if '–' in day_timing:
                    open_time, close_time = day_timing.split(' – ')
                    food_trucks.at[index, f'{day}_Open'] = open_time.strip()
                    food_trucks.at[index, f'{day}_Close'] = close_time.strip()

# Calculate the hours of operation for each day and sum them up for the total weekly hours
food_trucks['Total_Weekly_Hours'] = 0
for day in days:
    food_trucks[f'{day}_Hours'] = food_trucks.apply(lambda row: time_difference(row[f'{day}_Open'], row[f'{day}_Close']), axis=1)
    food_trucks['Total_Weekly_Hours'] += food_trucks[f'{day}_Hours']
# Drop the "Opening Hours" column
food_trucks = food_trucks.drop(columns=['Opening Hours'])
food_trucks.to_csv('food_trucks_cleaned.csv', index=False)
```

## Exploratory Data Analysis

```python
# Display basic statistics for the numeric columns (e.g., Rating)
print(food_trucks.describe())

# Basic Statistics of Ratings
mean_rating = food_trucks['Rating'].mean()
median_rating = food_trucks['Rating'].median()
std_dev_rating = food_trucks['Rating'].std()

print(f"Mean Rating: {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Standard Deviation of Ratings: {std_dev_rating:.2f}")


```

### 1. Distribution of Ratings Among Food Trucks

```python
# Distribution of ratings among the food trucks
plt.figure(figsize=(10, 6))
sns.histplot(food_trucks['Rating'], kde=True, bins=10)
plt.title('Distribution of Ratings Among Food Trucks')
plt.xlabel('Rating')
plt.ylabel('Number of Food Trucks')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### 2. Food Trucks with the highest total weekly hours

```python
# Food trucks with the highest total weekly hours
top_hours = food_trucks.sort_values(by='Total_Weekly_Hours', ascending=False).head()

fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Plotting the top hours
sns.barplot(data=top_hours, x='Name', y='Total_Weekly_Hours', ax=ax[0], palette='viridis')
ax[0].set_title('Top 5 Food Trucks with the Highest Weekly Operating Hours')
ax[0].set_ylabel('Total Weekly Hours')
ax[0].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

### 3. Distribution of Cuisine Types

```python
# Distribution of cuisine types
plt.figure(figsize=(12, 7))
sns.countplot(data=food_trucks, y='Cuisine Type', order=food_trucks['Cuisine Type'].value_counts().index, palette='viridis')
plt.title('Distribution of Cuisine Types Among Food Trucks')
plt.xlabel('Number of Food Trucks')
plt.ylabel('Cuisine Type')
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### 4. Average Operating Hours for Each Day of The Week Across all Food Trucks

```python
# Average operating hours for each day of the week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_hours = food_trucks[[day + '_Hours' for day in days_of_week]].mean()

plt.figure(figsize=(12, 7))
avg_hours.plot(kind='bar', color='teal')
plt.title('Average Operating Hours for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Operating Hours')
plt.xticks(rotation=45)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### 5. Correlation Between Ratings and Total Weekly Hours

```python
# Scatter plot to visualize the correlation between ratings and total weekly hours
plt.figure(figsize=(12, 7))
sns.scatterplot(data=food_trucks, x='Total_Weekly_Hours', y='Rating', hue='Rating', palette='viridis', size='Rating', sizes=(50, 150))
plt.title('Correlation Between Ratings and Total Weekly Hours')
plt.xlabel('Total Weekly Hours')
plt.ylabel('Rating')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
